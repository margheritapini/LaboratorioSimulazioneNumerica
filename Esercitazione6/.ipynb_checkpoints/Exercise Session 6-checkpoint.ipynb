````python
import numpy as np
import matplotlib.pyplot as plt
````

L'esercizio richiede di simulare un modello di Ising 1D per 50 spin con condizioni periodiche al contorno, utilizzando l'algoritmo di Metropolis e l'algoritmo di Gibbs.
Vengono studiati l'energia interna per particella $U/N$, la capacità termica $C$, la magnetizzazione $M$ e la suscettibilità $\chi$ a 11 diverse temperature $T$ selezionate equidistanti nell'intervallo [0.5, 2.5]. I risultati sono poi confrontati con i valori teorici dati dalla risoluzione esatta del modello.

Per ottenere le singole misure, sono partita dalla simulazione a $T=2.5$ e da una configurazione iniziale di spin randomica. Ho equilibrato il sistema con 2000 iterazioni ed ho poi ho utilizzato 20 blocchi da 20000 iterazioni ciascuno per effettuare le misure richieste. Al termine del ciclo di simulazione dedicato a tale temperatura, viene rieseguito il procedimento (compresa la fase di equilibrazione) con la temperatura immediatamente inferiore impostata tramite un apposito metodo della classe System:
``` cpp
void System :: change_temp(double decrease) {
  _average.zeros();
  _global_av.zeros();
  _global_av2.zeros();
  _nattempts = 0;
  _naccepted = 0;
  
  _temp -= decrease;
  _beta = 1/_temp;
}
```
In questo modo, per ogni nuova temperatura (inferiore alla precedente) la configurazione iniziale risulta essere la configurazione finale ottenuta dalla simulazione per la temperatura immediatamente precedente. Così fino ad arrivare alla temperatura finale $T=0.5$. In questo modo, la simulazione ad ogni temperatura inferiore ha automaticamente inizio da una configurazione non troppo distante dalla configurazione di equilibrio ricercata.

Per ogni quantità misurata, per tutte le temperature studiate sono mostrate le misure ottenute con i due procedimenti al termine di ciascuna simulazione e sono confrontate con la curva teorica. Per ogni misura, è stata poi valutata la discrepanza dal valore teorico:
$$
\frac{\text{valore misurato - valore teorico}}{\text{incertezza}_{\text{valore misurato}}}.
$$

Tutte le misure effettuate presentano una buona adesione con il modello teorico; vi è infatti un'unica misura con una discrepanza dal valore teorico maggiore di 2.

#### Energia interna

````python
# Define the parameters and theoretical model
points = 100
T = np.linspace(0.2, 3.0, num=points)
beta = 1 / T
J = 1.0
Ns = 50
th = np.tanh(J / T)
thN = th ** Ns
ch = 1 / th
e = -J * (th + ch * thN) / (1 + thN)

# Plot the theoretical model
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(T, e, color='black', linewidth=0.5, label='Theoretical model (h=0)')
plt.title('Internal energy')
plt.xlabel('T')
plt.ylabel('U/N')

# Read data points from file
file_path_mrt2 = './OUTPUT/MRT2_h0/total_energy_final.dat'
data_mrt2 = np.loadtxt(file_path_mrt2)

file_path_gibbs = './OUTPUT/GIBBS_h0/total_energy_final.dat'
data_gibbs = np.loadtxt(file_path_gibbs)

# Extract x, y, and error values from the data
x_values_mrt2 = data_mrt2[:, 0]
y_values_mrt2 = data_mrt2[:, 1]
error_values_mrt2 = data_mrt2[:, 2]

x_values_gibbs = data_gibbs[:, 0]
y_values_gibbs = data_gibbs[:, 1]
error_values_gibbs = data_gibbs[:, 2]

# Plot the data points from the file
plt.errorbar(x_values_mrt2, y_values_mrt2, yerr=error_values_mrt2, fmt='o', label='$M(RT)^2$', markerfacecolor='blue', markeredgecolor='none')
plt.errorbar(x_values_gibbs, y_values_gibbs, yerr=error_values_gibbs, fmt='^', label='GIBBS', markerfacecolor='orange', markeredgecolor='none')
plt.grid(True)
plt.legend()

# Calculate theoretical values at measured temperatures
def theoretical_energy(T):
    th = np.tanh(J / T)
    thN = th ** Ns
    ch = 1 / th
    return -J * (th + ch * thN) / (1 + thN)

# Compute (measured value - theoretical value) / error for MRT2 and GIBBS
theoretical= theoretical_energy(x_values_mrt2)
delta_mrt2 = (y_values_mrt2 - theoretical) / error_values_mrt2
delta_gibbs = (y_values_gibbs - theoretical) / error_values_gibbs

# Plot the delta values
plt.subplot(1, 2, 2)
plt.errorbar(x_values_mrt2, delta_mrt2, fmt='o', label='$M(RT)^2$', markerfacecolor='blue', markeredgecolor='none', linestyle='--', linewidth=0.5)
plt.errorbar(x_values_gibbs, delta_gibbs, fmt='o', label='GIBBS', markerfacecolor='orange', markeredgecolor='none', linestyle='--', linewidth=0.5)
plt.axhline(y=0, color='red', linestyle='--', linewidth=1)
plt.title('Deviation from Theoretical Model')
plt.xlabel('T')
plt.ylabel('$(\mathrm{measured} - \mathrm{theoretical}) / \mathrm{error}$')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
````

#### Capacità termica

````python
# Define the heat capacity theoretical model
heat = ((beta * J) ** 2) * (((1 + thN + (Ns - 1) * (th ** 2) + (Ns - 1) * (ch ** 2) * thN) / (1 + thN)) - Ns * ((th + ch * thN) / (1 + thN)) ** 2)

# Plot the heat capacity theoretical model
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(T, heat, color='black', linewidth=0.5, label='Theoretical model (h=0)')
plt.title('Heat capacity')
plt.xlabel('T')
plt.ylabel('C')

# Read data points from file
file_path_mrt2 = './OUTPUT/MRT2_h0/specific_heat_final.dat'
data_mrt2 = np.loadtxt(file_path_mrt2)

file_path_gibbs = './OUTPUT/GIBBS_h0/specific_heat_final.dat'
data_gibbs = np.loadtxt(file_path_gibbs)

# Extract x, y, and error values from the data
x_values_mrt2 = data_mrt2[:, 0]
y_values_mrt2 = data_mrt2[:, 1]
error_values_mrt2 = data_mrt2[:, 2]

x_values_gibbs = data_gibbs[:, 0]
y_values_gibbs = data_gibbs[:, 1]
error_values_gibbs = data_gibbs[:, 2]

# Plot the data points from the file
plt.errorbar(x_values_mrt2, y_values_mrt2, yerr=error_values_mrt2, fmt='o', label='$M(RT)^2$', markerfacecolor='blue', markeredgecolor='none')
plt.errorbar(x_values_gibbs, y_values_gibbs, yerr=error_values_gibbs, fmt='^', label='GIBBS', markerfacecolor='orange', markeredgecolor='none')
plt.grid(True)
plt.legend()

# Calculate theoretical values at measured temperatures
def theoretical_heat(T):
    beta = 1 / T
    J = 1.0
    Ns = 50
    th = np.tanh(J / T)
    thN = th ** Ns
    ch = 1 / th
    return ((beta * J) ** 2) * (((1 + thN + (Ns - 1) * (th ** 2) + (Ns - 1) * (ch ** 2) * thN) / (1 + thN)) - Ns * ((th + ch * thN) / (1 + thN)) ** 2)

# Compute (measured value - theoretical value) / error for MRT2 and GIBBS
theoretical_mrt2 = theoretical_heat(x_values_mrt2)
delta_mrt2 = (y_values_mrt2 - theoretical_mrt2) / error_values_mrt2
theoretical_gibbs = theoretical_heat(x_values_gibbs)
delta_gibbs = (y_values_gibbs - theoretical_gibbs) / error_values_gibbs

# Plot the delta values
plt.subplot(1, 2, 2)
plt.errorbar(x_values_mrt2, delta_mrt2, fmt='o', label='$M(RT)^2$', markerfacecolor='blue', markeredgecolor='none', linestyle='--', linewidth=0.5)
plt.errorbar(x_values_gibbs, delta_gibbs, fmt='^', label='GIBBS', markerfacecolor='orange', markeredgecolor='none', linestyle='--', linewidth=0.5)
plt.axhline(y=0, color='red', linestyle='--', linewidth=0.6)
plt.title('Deviation from Theoretical Model')
plt.xlabel('T')
plt.ylabel('$(\mathrm{measured} - \mathrm{theoretical}) / \mathrm{error}$')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
````

#### Magnetizzazione

````python
# Define the parameters and theoretical model
points = 100
T = np.linspace(0.2, 3.0, num=points)
J = 1.0
Ns = 50
h = 0.02  # external field
b = 1 / T

l1 = np.exp(b * J) * np.cosh(b * h) + np.sqrt(np.exp(2 * b * J) * np.cosh(b * h) * np.cosh(b * h) - 2 * np.sinh(2 * b * J))
l2 = np.exp(b * J) * np.cosh(b * h) - np.sqrt(np.exp(2 * b * J) * np.cosh(b * h) * np.cosh(b * h) - 2 * np.sinh(2 * b * J))
Z = l1 ** Ns + l2 ** Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
# Plot the theoretical model
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(T, M, color='black', linewidth=0.5, label='Theoretical model (h=0.02)')
plt.title('Magnetization M with h = 0.02')
plt.xlabel('T')
plt.ylabel('$M$')

# Read data points from file
file_path_mrt2 = './OUTPUT/MRT2_h.02/magnetization_final.dat'
data_mrt2 = np.loadtxt(file_path_mrt2)

file_path_gibbs = './OUTPUT/GIBBS_h.02/magnetization_final.dat'
data_gibbs = np.loadtxt(file_path_gibbs)

# Extract x, y, and error values from the data
x_values_mrt2 = data_mrt2[:, 0]
y_values_mrt2 = data_mrt2[:, 1]
error_values_mrt2 = data_mrt2[:, 2]

x_values_gibbs = data_gibbs[:, 0]
y_values_gibbs = data_gibbs[:, 1]
error_values_gibbs = data_gibbs[:, 2]

# Plot the data points from the file
plt.errorbar(x_values_mrt2, y_values_mrt2, yerr=error_values_mrt2, fmt='o', label='$M(RT)^2$', markerfacecolor='blue', markeredgecolor='none')
plt.errorbar(x_values_gibbs, y_values_gibbs, yerr=error_values_gibbs, fmt='^', label='GIBBS', markerfacecolor='orange', markeredgecolor='none')
plt.grid(True)
plt.legend()

# Calculate theoretical values at measured temperatures
def theoretical_magnetization(T):
    b = 1 / T
    l1 = np.exp(b * J) * np.cosh(b * h) + np.sqrt(np.exp(2 * b * J) * np.cosh(b * h) * np.cosh(b * h) - 2 * np.sinh(2 * b * J))
    l2 = np.exp(b * J) * np.cosh(b * h) - np.sqrt(np.exp(2 * b * J) * np.cosh(b * h) * np.cosh(b * h) - 2 * np.sinh(2 * b * J))
    Z = l1 ** Ns + l2 ** Ns
    M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)    
    return M

# Compute (measured value - theoretical value) / error for MRT2 and GIBBS
theoretical_mrt2 = theoretical_magnetization(x_values_mrt2)
delta_mrt2 = (y_values_mrt2 - theoretical_mrt2) / error_values_mrt2

theoretical_gibbs = theoretical_magnetization(x_values_gibbs)
delta_gibbs = (y_values_gibbs - theoretical_gibbs) / error_values_gibbs

# Plot the delta values
plt.subplot(1, 2, 2)
plt.errorbar(x_values_mrt2, delta_mrt2, fmt='o', label='$M(RT)^2$', markerfacecolor='blue', markeredgecolor='none', linestyle='-', linewidth=0.5)
plt.errorbar(x_values_gibbs, delta_gibbs, fmt='^', label='GIBBS', markerfacecolor='orange', markeredgecolor='none', linestyle='-', linewidth=0.5)
plt.axhline(y=0, color='red', linestyle='--', linewidth=0.6)
plt.title('Deviation from Theoretical Model')
plt.xlabel('T')
plt.ylabel('$(\mathrm{measured} - \mathrm{theoretical}) / \mathrm{error}$')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
````

#### Suscettività

````python
# Define the parameters and theoretical model
points = 100
T = np.linspace(0.2, 3.0, num=points)
J = 1.0
Ns = 50
beta = 1 / T
th = np.tanh(J / T)
thN = th ** Ns
X = beta * np.exp(2 * beta * J) * (1 - thN) / (1 + thN)

# Plot the theoretical model for susceptibility
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(T, X, color='black', linewidth=0.5, label='Theoretical model (h=0)')
plt.title('Susceptibility')
plt.xlabel('T')
plt.ylabel('$\chi$')

# Read data points from file
file_path_mrt2 = './OUTPUT/MRT2_h0/susceptibility_final.dat'
data_mrt2 = np.loadtxt(file_path_mrt2)

file_path_gibbs = './OUTPUT/GIBBS_h0/susceptibility_final.dat'
data_gibbs = np.loadtxt(file_path_gibbs)

# Extract x, y, and error values from the data
x_values_mrt2 = data_mrt2[:, 0]
y_values_mrt2 = data_mrt2[:, 1]
error_values_mrt2 = data_mrt2[:, 2]

x_values_gibbs = data_gibbs[:, 0]
y_values_gibbs = data_gibbs[:, 1]
error_values_gibbs = data_gibbs[:, 2]

# Plot the data points from the file without connecting lines
plt.errorbar(x_values_mrt2, y_values_mrt2, yerr=error_values_mrt2, fmt='o', label='$M(RT)^2$', markerfacecolor='blue', markeredgecolor='none')
plt.errorbar(x_values_gibbs, y_values_gibbs, yerr=error_values_gibbs, fmt='^', label='GIBBS', markerfacecolor='orange', markeredgecolor='none')
plt.grid(True)
plt.legend()

# Calculate theoretical values at measured temperatures
def theoretical_susceptibility(T):
    beta = 1 / T
    th = np.tanh(J / T)
    thN = th ** Ns
    return beta * np.exp(2 * beta * J) * (1 - thN) / (1 + thN)

# Compute (measured value - theoretical value) / error for MRT2 and GIBBS
theoretical_mrt2 = theoretical_susceptibility(x_values_mrt2)
delta_mrt2 = (y_values_mrt2 - theoretical_mrt2) / error_values_mrt2

theoretical_gibbs = theoretical_susceptibility(x_values_gibbs)
delta_gibbs = (y_values_gibbs - theoretical_gibbs) / error_values_gibbs

# Plot the delta values
plt.subplot(1, 2, 2)
plt.errorbar(x_values_mrt2, delta_mrt2, fmt='o', label='$M(RT)^2$', markerfacecolor='blue', markeredgecolor='none', linestyle='-', linewidth=0.5)
plt.errorbar(x_values_gibbs, delta_gibbs, fmt='^', label='GIBBS', markerfacecolor='orange', markeredgecolor='none', linestyle='-', linewidth=0.5)
plt.axhline(y=0, color='red', linestyle='--', linewidth=0.6)
plt.title('Deviation from Theoretical Model')
plt.xlabel('T')
plt.ylabel('$(\mathrm{measured} - \mathrm{theoretical}) / \mathrm{error}$')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
````
