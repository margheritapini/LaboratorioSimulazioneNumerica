```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

La teoria di Black-Scholes assume che il prezzo degli asset $S$ sia un processo stocastico e che segua un moto Browniano geometrico (GBM) con tasso di interesse costante $r$, privo di rischio, e volatilit√† $\sigma$. 

Fissati i parametri ($S(0)=100$, $T=1$, $K=100$, $r=0.1$, $\sigma=0.25$), l'esercizio richiede di calcolare al tempo $t=0$, attraverso stima Monte Carlo, il prezzo di un'opzione-call europea $C[S(0),0]$ ed il prezzo di un'opzione-put europea $P[S(0),0]$.


##### Stima diretta

$S\sim GBM(\mu,\sigma^2)$, allora
    $$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$
    dove $W(t)\sim\mathcal{N}(0,t)$.

Ho effettuato la stima $M=10^7$ volte, suddivise in $N=100$ blocchi. Per effettuare la singola stima, ho estratto $Z$ da una distribuzione gaussiana $Z \sim \mathcal{N}(0,1)$, che ho poi utilizzato per valutare $S(T)$. A partire da tale valore, ad ogni interazione ho calcolato $call$ e $put$ come segue:
- call = $exp(-rT) \cdot max(0, S-K)$
- put = $exp(-rT) \cdot max(0, K-S)$

Ho confrontato i valori ottenuti con le soluzioni analitiche delle equazioni di Black-Scholes con i parametri forniti. Le misure ottenute con la simulazione sono compatibili con il valore teorico.

```python
# Load data from file for the first plot
k, call_dir, error_call_dir, put_dir, error_put_dir = np.loadtxt("call_put_direct.dat", usecols=(0, 1, 2, 3, 4), delimiter=' ', unpack=True)

# Create subplots with 1 row and 2 columns
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Plot the first graph on the left subplot
axs[0].errorbar(k, call_dir, yerr=error_call_dir, fmt='o', markersize=2, color='black', ecolor='black', linewidth=0.5)
axs[0].set_xlabel('#steps')
axs[0].set_ylabel('$C[S(0),0]$')
axs[0].axhline(y=14.975790778311286, color='red', linewidth=0.9, label='Black-Scholes solution')
axs[0].grid(True)
axs[0].legend()

# Plot the second graph on the right subplot
axs[1].errorbar(k, put_dir, yerr=error_put_dir,  fmt='o', markersize=2, color='black', ecolor='black', linewidth=0.4)
axs[1].set_xlabel('#steps')
axs[1].set_ylabel('$P[S(0),0]$')
axs[1].axhline(y=5.4595325819072364, color='red', linewidth=0.9, label='Black-Scholes solution')
axs[1].grid(True)
axs[1].legend()

fig.suptitle('Sampling directly the final asset price', fontsize=16)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

##### Stima discretizzata

Ho seguito lo stesso procedimento seguito nella "stima diretta", ma ottenendo $S(T)$ ricorsivamente.
Ho suddiviso l'intervallo [0,T] in 100 sotto-intervalli identici ed ho valutato $S(t_{i+1})$ a partire da $S(t_{i})$:

$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu - \frac{1}{2}\sigma^2\right)(t_{i+1} - t_i) + \sigma Z_{i+1}\sqrt{t_{i+1} - t_i}\right]$

sfruttando il fatto che gli incrementi di $W$ sono indipendenti e distribuiti normalmente.

Anche in questo caso, ho confrontato i valori ottenuti con le soluzioni analitiche delle equazioni di Black-Scholes con i parametri forniti e le misure ottenute con la simulazione sono compatibili con il valore teorico.

```python
# Load data from file for the first plot
k, call_dir, error_call_dir, put_dir, error_put_dir = np.loadtxt("call_put_discretized.dat", usecols=(0, 1, 2, 3, 4), delimiter=' ', unpack=True)

# Create subplots with 1 row and 2 columns
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# Plot the first graph on the left subplot
axs[0].errorbar(k, call_dir, yerr=error_call_dir, fmt='o', markersize=2, color='black', ecolor='black', linewidth=0.5)
axs[0].set_xlabel('#steps')
axs[0].set_ylabel('$C[S(0),0]$')
axs[0].axhline(y=14.975790778311286, color='red', linewidth=0.9, label='Black-Scholes solution')
axs[0].grid(True)
axs[0].legend()

# Plot the second graph on the right subplot
axs[1].errorbar(k, put_dir, yerr=error_put_dir,  fmt='o', markersize=2, color='black', ecolor='black', linewidth=0.4)
axs[1].set_xlabel('#steps')
axs[1].set_ylabel('$P[S(0),0]$')
axs[1].axhline(y=5.4595325819072364, color='red', linewidth=0.9, label='Black-Scholes solution')
axs[1].grid(True)
axs[1].legend()

fig.suptitle('Sampling the discretized path of the asset prize', fontsize=16)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
