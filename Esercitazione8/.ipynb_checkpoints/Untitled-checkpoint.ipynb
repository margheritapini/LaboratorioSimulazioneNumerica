```python
import numpy as np
import matplotlib.pyplot as plt

# Parametri
mu = 0.8
sigma = 0.6

# Definisci la funzione Psi_T
def psi_T(x, mu, sigma):
    return np.exp(-(x - mu)**2 / (2 * sigma**2)) + np.exp(-(x + mu)**2 / (2 * sigma**2))

# Genera un intervallo di valori per x
x_values = np.linspace(-2, 2, 400)

# Calcola i valori della funzione per l'intervallo di x
y_values = psi_T(x_values, mu, sigma)

# Calcola il modulo quadro della funzione
modulo_quadro = np.abs(y_values)**2

# Disegna il grafico del modulo quadro
plt.plot(x_values, modulo_quadro, label=r'$|\Psi_T^{\sigma,\mu}(x)|^2$')
plt.title(r'$|\Psi_T^{\sigma,\mu}(x)|^2$ con $\mu = 0.8$ e $\sigma = 0.6$')
plt.xlabel('x')
plt.ylabel(r'$|\Psi_T^{\sigma,\mu}(x)|^2$')
plt.legend()
plt.grid(True)

# Mostra il grafico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Leggere i dati dal file H_SA.dat
data = np.loadtxt('H_SA.dat')

# Estrarre le colonne
x = data[:, 0]  # Prima colonna
y = data[:, 1]  # Seconda colonna
yerr = data[:, 2]  # Terza colonna

plt.figure(figsize=(12, 8))


# Creare il grafico
plt.errorbar(x[400:], y[400:], yerr=yerr[400:], fmt='o', ecolor='r', capsize=0.1, label='Dati con errore', markersize=2)
plt.axhline(-0.46046587969271513)
plt.xlabel('Ascissa (Prima colonna)')
plt.ylabel('Ordinata (Seconda colonna)')
plt.title('Grafico dei dati con errore')
plt.legend()
plt.grid(True)

# Mostrare il grafico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Leggere i dati dal file H_SA.dat
data = np.loadtxt('H_SA.dat')

# Estrarre le colonne
x = data[:, 3]  # Quarta colonna
y = data[:, 4]  # Quinta colonna

# Creare la figura con dimensioni personalizzate
plt.figure(figsize=(12, 8))

# Tracciare il percorso dei punti
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Percorso dei punti', markersize=0.1, linewidth=0.1)

# Evidenziare l'ultimo punto con una stellina
plt.plot(x[-1], y[-1], marker='*', color='r', markersize=5, label='Ultimo punto (mu, sigma)')

# Aggiungere etichette e titolo
plt.xlabel('Quarta colonna')
plt.ylabel('Quinta colonna')
plt.title('Percorso dei punti definiti da (quarta colonna, quinta colonna)')
plt.legend()
plt.grid(True)

# Mostrare il grafico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Leggi i dati dal file
file_path = 'paolo.dat'
x = []
y = []

with open(file_path, 'r') as file:
    for line in file:
        columns = line.strip().split(',')
        x.append(float(columns[0]))
        y.append(float(columns[1]))

# Crea il grafico
plt.figure(figsize=(10, 6))
plt.plot(x[100:], y[100:], marker='o', linestyle='-', color='b')
plt.xlabel('Ascissa (Prima Colonna)')
plt.ylabel('Ordinata (Seconda Colonna)')
plt.title('Grafico dei dati da paolo.dat')
plt.grid(True)
plt.show()
```
