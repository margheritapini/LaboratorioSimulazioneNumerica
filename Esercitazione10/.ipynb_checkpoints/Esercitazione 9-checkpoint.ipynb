```python
# Carica i dati da "Misure.dat"
data = np.loadtxt("Misure_circonferenza.dat")

# Estrai le colonne di interesse
colonna1 = data[:200, 0]
colonna2 = data[:200, 1]
colonna3 = data[:200, 2]

# Crea la figura e i subplot orizzontali
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot del primo subplot
ax1.plot(colonna1, colonna2, marker='o', markersize=2, linestyle='', color='black')
ax1.set_title('')
ax1.set_xlabel('Generazione')
ax1.set_ylabel('$L_{best}$')

# Plot del secondo subplot
ax2.plot(colonna1, colonna3, marker='o', color='black', linestyle='', markersize=2)
ax2.set_title('')
ax2.set_xlabel('Generazione')
ax2.set_ylabel('$<L>_{best population}$')

# Mostra il grafico
plt.tight_layout()
plt.show()
```
