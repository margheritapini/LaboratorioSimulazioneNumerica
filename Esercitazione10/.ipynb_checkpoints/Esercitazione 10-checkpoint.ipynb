````python
import matplotlib.pyplot as plt
import numpy as np
````

L'esercizio richiede di parallelizzare con librerie MPI il codice genetico sviluppato nell'Esercitazione 9 e di applicarlo ad un percorso di 110 città.

Ho sviluppato il codice come segue:
- Ciascun core, con il proprio seme del Random Generator, sviluppa la propria ricerca indipendente (parallelizzazione del tipo SPMD);
- Ogni $N_{migr}$ generazioni vengono estratte coppie casuali di core e ciascuna coppia si scambia l'individuo migliore.

Rispetto al codice precedente ho modificato:
- Il numero di membri totale della popolazione: $\_npopulation=10000$, per ottenere maggior diversificazione;
- L'operatore di selezione, per adattarlo meglio alla popolazione più numerosa, solo nell'esponente:
  $$
\text{int}\left(\frac{\_npopulation-1}{0.5^\textcolor{red}{8}} \cdot (r-0.5)^{\textcolor{red}{8}} \right), \quad r \in(0,1) \text{ random}
$$

Ho eseguito il codice su 8 core. È possibile eseguirlo su più core, purchè in numero pari, tramite:
```bash 
make
mpirun -np cores_number ./main.exe

````python
# Coordinate delle città
data = np.loadtxt("cap_prov_ita.dat")

coordinate_x = data[:, 0]
coordinate_y = data[:, 1]
````

#### Core indipendenti

Ho prima eseguito il codice su otto core indipendenti, senza nessuna comunicazione, per 3000 generazioni. I processi non giungono alla medesima soluzione. I percorsi trovati sono comunque buoni (non presentano intersezioni e spesso presentano porzioni di percorso simili), ma sono molto probabilmente minimi locali.

````python
# Creazione dei subplot per i grafici
fig, axs = plt.subplots(2, 4, figsize=(25, 12))
axs = axs.flatten()

for i in range(8):
    filename = f"INDEPENDENT_CORES/best_path{i}.dat"
    data = np.loadtxt(filename)
    order = data[:110]
    distance = data[-1]
    order = np.append(order, order[0])
    x_coords_ordered = [coordinate_x[int(i - 1)] for i in order]
    y_coords_ordered = [coordinate_y[int(i - 1)] for i in order]

    ax = axs[i]
    ax.plot(x_coords_ordered, y_coords_ordered, marker='o', linestyle='-', color='black', linewidth=0.5, markersize=1)
    ax.set_title(f"Distance: {distance}", fontsize=20)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')

plt.tight_layout()
plt.show()
````

Ho graficato l'andamento della distanza $L_{best}$ del percorso migliore per ciascun processo. Ciascun processo, in circa 2500 generazioni, si stabilizza sulla propria proposta di percorso (anche se non necessariamente in modo definitivo: si veda ad esempio il processo 1, in rosa, che dopo circa 2000 generazioni di stazionarietà propone un percorso migliore). 

Tre degli otto processi hanno ottenuto un percorso con distanza complessiva inferiore a 66.

````python
# Plot completo
plt.figure(figsize=(10, 4))

for i in range(8):
    filename = f"INDEPENDENT_CORES/Best_distance{i}.dat"
    data = np.loadtxt(filename)
    generation = data[:,0]
    best_L = data[:,1]

    filename_path = f"INDEPENDENT_CORES/best_path{i}.dat"
    data = np.loadtxt(filename_path)
    distance = data[-1]

    plt.plot(generation, best_L, marker='o', markersize=0.5, linestyle='', lw=0.5, label=f"Core {i} - Distance: {distance}")

plt.title('Evoluzione di $L_{best}$ nelle diverse prove')
plt.xlabel('Generazione')
plt.ylabel('$L_{best}$')
plt.legend()
plt.tight_layout()
plt.show()

# Focus su sezione con y-limit
plt.figure(figsize=(10, 5))

for i in range(8):
    filename = f"INDEPENDENT_CORES/Best_distance{i}.dat"
    data = np.loadtxt(filename)
    generation = data[:,0]
    best_L = data[:,1]

    filename_path = f"INDEPENDENT_CORES/best_path{i}.dat"
    data = np.loadtxt(filename_path)
    distance = data[-1]

    plt.plot(generation, best_L, marker='o', markersize=0.5, linestyle='', lw=0.5,label=f"Core {i} - Distance: {distance}")

plt.title('Evoluzione di $L_{best}$ nelle diverse prove (Zoom)')
plt.xlabel('Generazione')
plt.ylabel('$L_{best}$')
plt.ylim(62.5, 80)
plt.legend()
plt.tight_layout()
plt.show()
````

Dopo circa 500 generazioni, l'algoritmo si stabilizza su di un unico percorso che è più corto rispetto a tutti i percorsi trovati nella simulazione eseguita a core indipendenti. Questo percorso ha una lunghezza simile a quella ottenuta solo da tre degli otto core indipendenti.

Per chiarezza, riporto sullo stesso plot l'andamento di $L_{best}$ al variare delle generazioni nelle due diverse simulazioni.

````python
plt.figure(figsize=(12, 6))

# Plot dei dati COMMUNICATING_CORES
for i in range(8):
    filename = f"COMMUNICATING_CORES/20/Best_distance{i}.dat"
    data = np.loadtxt(filename)
    generation = data[:, 0]
    best_L = data[:, 1]

    filename_path = f"COMMUNICATING_CORES/20/best_path{i}.dat"
    data = np.loadtxt(filename_path)
    distance = data[-1]

    plt.plot(generation, best_L, marker='o', markersize=1, linestyle='-', label=f"Comm. Core {i} - Distance: {distance}")

# Plot dei dati INDEPENDENT_CORES
for i in range(8):
    filename = f"INDEPENDENT_CORES/Best_distance{i}.dat"
    data = np.loadtxt(filename)
    generation = data[:, 0]
    best_L = data[:, 1]

    filename_path = f"INDEPENDENT_CORES/best_path{i}.dat"
    data = np.loadtxt(filename_path)
    distance = data[-1]

    plt.plot(generation, best_L, marker='x', markersize=1, label=f"Indep. Core {i} - Distance: {distance}", alpha=0.08)

plt.title('Evoluzione di $L_{best}$ nelle diverse prove (Zoom)')
plt.xlabel('Generazione')
plt.ylabel('$L_{best}$')
plt.ylim(65, 80)
plt.legend()
plt.tight_layout()
plt.show()
````

Nonostante i diversi processi si scambino un unico individuo, l'effetto complessivo della parallelizzazione è decisamente pronunciato. In circa 500 generazioni, il codice parallelizzato si stabilizza su di una buona proposta di percorso. 

Il codice fatica però ad uscire dai minimi locali: per ben 2500 generazioni, dopo aver raggiunto il minimo locale 65.26, non ci sono miglioramenti. Per ovviare a questo difetto si potrebbe provare ad introdurre delle nuove mutazioni, o ad aumentare la probabilità delle stesse (magari all'avanzare delle generazioni, quando la popolazione tende ad uniformarsi sul minimo locale ed è necessario rinforzare la diversificazione).

In ogni caso, la parallelizzazione ha sicuramente permesso di raggiungere valori minimi che poco probabilmente sono ottenuti da un processo indipendente, riducendo notevolmente il numero di generazioni necessarie alla convergenza.

#### Modifica: sensibile aumento della probabilità di mutazioni

Ho aumentato sensibilmente la probabilità delle mutazioni ed effettivamente ho ottenuto un miglioramento del risultato!
Molto probabilmente, agendo ancora pù finemente sulle probabilità di mutazione, è possibile rendere l'algoritmo più agile ad uscire dai minimi locali.
